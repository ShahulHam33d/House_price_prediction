### Missing values

#### -Categorical features

```python
# Handeling nan values in categorical features
# Here "tr" refers to train and "tt" refers to test
tr_feature_nan = [feature for feature in df_train.columns if df_train[feature].isnull().sum()>0 and 
                  df_train[feature].dtype == "O"]
tt_feature_nan = [feature for feature in df_test.columns if df_test[feature].isnull().sum()>0 and 
                  df_test[feature].dtype == "O"]
print("train_data:-")
print("\nno of features: ",len(tr_feature_nan))
for feature in tr_feature_nan:
    print("{} : {} % missing values".format(feature,np.round(df_train[feature].isnull().mean()*100,2)))

print("\ntest_data:-")
print("\nno of features: ",len(tt_feature_nan))
for feature in tt_feature_nan:
    print("{} : {} % missing values".format(feature,np.round(df_test[feature].isnull().mean(),4)))
    
```

```python
def replace_nan_features(dataset,features_nan):
    data = dataset.copy()
    data[features_nan] = data[features_nan].fillna("missing")
    return data

df_train = replace_nan_features(df_train,tr_feature_nan)
df_test = replace_nan_features(df_test,tt_feature_nan)
print("train data:-\n",df_train[tr_feature_nan].isnull().sum())
print("")
print("test data:-\n",df_test[tt_feature_nan].isnull().sum())
```

```python
df_train.head()
```

```python
# now lets work with numerical values
# note: tr means train data and tt means test data

tr_num_nan = [feature for feature in df_train.columns if df_train[feature].isnull().sum()>0 
              and df_train[feature].dtype != "O"]
print("train data:-")
for feature in tr_num_nan:
    print("{} : {} % missing values".format(feature,np.round(df_train[feature].isnull().mean(),4)))
print("")

tt_num_nan = [feature for feature in df_test.columns if df_test[feature].isnull().sum()>0 
              and df_test[feature].dtype != "O"]
print("test data:-")
for feature in tt_num_nan:
    print("{} : {} % missing values".format(feature,np.round(df_test[feature].isnull().mean(),4)))
```

```python
#replacing numerical missing values with median since it has outliers
for feature in tr_num_nan:
    median_value = df_train[feature].median()
    
    # now lets create a new feature for better understanding of nan values
    df_train[feature+"_nan"] = np.where(df_train[feature].isnull(),1,0)
    df_train[feature] = df_train[feature].fillna(median_value)
print("train_data:-")
print(df_train[tr_num_nan].isnull().sum())
print()

for feature in tt_num_nan:
    median_value = df_test[feature].median()
    
    # now lets create a new feature for better understanding of nan values
    df_test[feature+"_nan"] = np.where(df_test[feature].isnull(),1,0)
    df_test[feature] = df_test[feature].fillna(median_value)
print("test_data:-")
print(df_test[tt_num_nan].isnull().sum())
```

```python
# Lets balance the dimension of both the data set
# Because the nmbr of features in train and test dataset should b equal

not_in_train = [feature for feature in df_test.columns if feature not in df_train.columns]
not_in_test = [feature for feature in df_train.columns if feature not in df_test.columns]

print("train data: ",not_in_train)
print("\ntest data: ",not_in_test)
```

```python
# in test dataset only the saleprice is missing which is not neccesary
# we will add the new columns to train data which we found on test dataset and fill with 0(since itr has no null value)

for feature in not_in_train:
    df_train[feature] = 0
```

#### - Numerical values

since numerical features are skewed i will perform log normal distribution

```python
#recall EDA applying log normal distribution on features which has no 0 values in it
temp = [feature for feature in df_train.columns if df_train[feature].dtype!="O" 
                and 0 not in df_train[feature].unique() and len(df_train[feature].unique()) > 25]
print(df_train[temp].head())

temp2 = [feature for feature in df_test.columns if df_test[feature].dtype!="O" 
                and 0 not in df_test[feature].unique() and len(df_test[feature].unique()) > 25]
print(df_test[temp2].head())
```

```python
tr_num_features = [feature for feature in temp if feature not in ["Id","SalePrice"]]
for feature in tr_num_features:
    df_train[feature] = np.log(df_train[feature])
    
tt_num_features = [feature for feature in temp2 if feature != "Id"]
for feature in tt_num_features:
    df_test[feature] = np.log(df_test[feature])
```

### Handling rare categorical features
i will remove categorical variables less than 1% in that feature

```python
tr_cat_features = [feature for feature in df_train.columns if df_train[feature].dtype == "O"]
tt_cat_features = [feature for feature in df_test.columns if df_test[feature].dtype == "O"]
print("no of columns: ",len(tr_cat_features))
df_train[tr_cat_features].head()
```

```python
for feature in tr_cat_features:
    temp = df_train.groupby(feature)["SalePrice"].count()/len(df_train)
    temp_df = temp[temp > 0.01].index # only indices which contains less than 1% are assigned to tem_df
    df_train[feature] = np.where(df_train[feature].isin(temp_df),df_train[feature],"rare_var")
df_train[tr_cat_features].head(5)
```

```python
for feature in tt_cat_features:
    not_rare = [var for var in df_train[feature].unique() if var != "rare_var" ]
    df_test[feature] = np.where(df_test[feature].isin(not_rare),df_test[feature],"rare_var")
df_test[tt_cat_features].sample(10)
```

```python
#checking
print("train data: ",len(df_train.columns))
print("test data: ",len(df_test.columns))# it will be one less bcoz of sale price column
```

### Feature Scaling



 let us separate categorical features and numerical features for feature scaling.
 
 I will be not performing feature scaling on nan features(which i have created) and id and saleprice feature also.

```python
tr_nan = [ feature for feature in df_train.columns if "nan" in feature]
tt_nan = [feature for feature in df_test.columns if "nan" in feature]

tr_cat = [feature for feature in df_train.columns if df_train[feature].dtype == "O"]
tt_cat = [feature for feature in df_test.columns if df_test[feature].dtype == "O"]

temp1 = tr_nan+tr_cat+["Id","SalePrice"]
temp2 = tt_nan+tt_cat+["Id"]

tr_num = [feature for feature in df_train.columns if feature not in temp1]
tt_num = [feature for feature in df_test.columns if feature not in temp2]
len(tr_cat)
```

#### numerical Scaling

```python
#feature scaling on numerical features
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()

df_train[tr_num] = mms.fit_transform(df_train[tr_num])
df_test[tt_num] = mms.fit_transform(df_test[tt_num])
```

```python
#checking
print("train data: ",len(df_train.columns))
print("test data: ",len(df_test.columns))# it will be one less bcoz of sale price column
```

#### categorical scaling

```python
# feature scaling on Categorical features
#train data
tr_cat_miss = [feature for feature in tr_cat if "missing" in df_train[feature].unique()]
tr_cat_rar = [feature for feature in tr_cat if "rare_var" in df_train[feature].unique()]
for feature in tr_cat_miss:
    df_train[feature+"_miss"] = np.where(df_train[feature]=="missing",1,0) 
for feature in tr_cat_rar:
    df_train[feature+"_rar"] = np.where(df_train[feature]=="rare_var",1,0)
    

#test data
tt_cat_miss = [feature for feature in tt_cat if "missing" in df_test[feature].unique()]
tt_cat_rar = [feature for feature in tt_cat if "rare_var" in df_test[feature].unique()]
for feature in tt_cat_miss:
    df_test[feature+"_miss"] = np.where(df_test[feature]=="missing",1,0) 
for feature in tt_cat_rar:
    df_test[feature+"_rar"] = np.where(df_test[feature]=="rare_var",1,0)
```
